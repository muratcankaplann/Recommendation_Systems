```python
def create_user_movie_df(limit =1000):
    import pandas as pd
    movie = pd.read_csv('datasets/movie_lens_dataset/movie.csv')
    rating = pd.read_csv('datasets/movie_lens_dataset/rating.csv')
    df = movie.merge(rating, how="left", on="movieId")
    comment_counts = pd.DataFrame(df["title"].value_counts())
    rare_movies = comment_counts[comment_counts["title"] <= limit].index
    common_movies = df[~df["title"].isin(rare_movies)]
    user_movie_df = common_movies.pivot_table(index=["userId"], columns=["title"], values="rating")
    return user_movie_df
```

```python
random_user = int(user_ratings.sample(1,random_state = 45).index.values[0])
```

```python
random_user_df = user_ratings[user_ratings.index == random_user]
```

```python
movie_watched = (random_user_df.loc[:,random_user_df.notna().any()].columns).tolist()
```

```python
movies_watched_df = user_ratings.loc[:,user_ratings.columns.isin(movie_watched)]
```

```python
movies_watched_df
```

```python
user_movie_count = movies_watched_df.T.notnull().sum()
```

```python
type(user_movie_count)
```

```python
user_movie_count = user_movie_count.reset_index()
```

```python
user_movie_count.columns = ["userId", "movie_count"]
```

```python
users_same_movies = user_movie_count[user_movie_count["movie_count"] >= (len(movie_watched)*0.6)]
```

```python
same_movie_user_id = users_same_movies["userId"].values.tolist()
```

```python
movies_watched_df_T = movies_watched_df.T
```

```python
same_movie_users_df = movies_watched_df_T.loc[:,movies_watched_df_T.columns.isin(same_movie_user_id)]
```

```python
corr_df = same_movie_users_df.corr()
```

```python
corr_df_random_user = corr_df.loc[:,random_user_df.index[0]].reset_index()
```

```python
corr_df_random_user.columns = ["userId", "correlation"]
```

```python
top_users_df = corr_df_random_user[corr_df_random_user["correlation"] >= 0.65]
```

```python
top_users_df = top_users_df.merge(rating_df,how = "inner", on = "userId")
```

```python
top_users_df = top_users_df[~(top_users_df["userId"] == random_user_df.index[0])]
```

```python
top_users_df["weighted_rating"] = top_users_df["correlation"] * top_users_df["rating"]
```

```python
top_users_df
```

```python
recommendation_df = top_users_df.groupby("movieId").agg({"weighted_rating": "mean"})
```

```python
recommendation_df = recommendation_df[recommendation_df["weighted_rating"]>=3.5].sort_values(by= "weighted_rating",ascending = False)
```

```python
recommendation_list_ID = recommendation_df.index.tolist()
```

```python
recommendation_list_ID= list(set(recommendation_list_ID) - set(movie_watched))
```

```python
recommendation_list_ID
```

```python
recommendation_list_name = movie_df.loc[movie_df["movieId"].isin(recommendation_list_ID),"title"].values.tolist()
```
