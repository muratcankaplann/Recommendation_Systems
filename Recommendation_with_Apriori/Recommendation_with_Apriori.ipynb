```python
def string_combination(x,y):
    z = 0
    x = str(x)
    y = str(y)
    z = x + "_" + y
    return z
```

```python
df["Hizmet"] = df.apply(lambda x: string_combination(x["ServiceId"],x["CategoryId"]), axis = 1)
```

```python
df.dtypes
```

```python
import datetime
```

```python
df['CreateDate'] = pd.to_datetime(df['CreateDate'])
```

```python
df
```

```python
def year_month (x,y):
    y = str(y)
    x = str(x)
    year = x[0:4]
    month = x[5:7]
    result = y + "_" + year + "_" + month
    return result
    
```

```python
df["SepetID"] = df.apply(lambda x: year_month(x["CreateDate"],x["UserId"]), axis = 1)
```

```python
df
```

```python
hizmet_sepet_df = df.pivot_table(values = ["UserId"], index=['SepetID'], columns=["Hizmet"],aggfunc='count')
```

```python
hizmet_sepet_df 
```

```python
hizmet_sepet_df.shape
```

```python
hizmet_sepet_df = hizmet_sepet_df.fillna(0)
```

```python
hizmet_sepet_df
```

```python
hizmet_sepet_df = hizmet_sepet_df.applymap(lambda x: 1 if x>0 else 0)
```

```python
hizmet_sepet_df = hizmet_sepet_df.droplevel(0, axis = 1)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_itemsets = apriori(hizmet_sepet_df,
                            min_support=0.01,
                            use_colnames=True)
```

```python
frequent_itemsets
```

```python
hizmet_sepet_df[hizmet_sepet_df["0_8"] == 1].shape
```

```python
frequent_itemsets.sort_values("support", ascending=False)
```

```python
rules = association_rules(frequent_itemsets,
                          metric="support",
                          min_threshold=0.01)
```

```python
rules
```

```python
rules["antecedents"] = rules["antecedents"].apply(lambda x: list(x)[0]).astype("unicode")
rules["consequents"] = rules["consequents"].apply(lambda x: list(x)[0]).astype("unicode")
```

```python
def arl_recommender(rules_dataframe, product , rec_count = 1) :
    sorted_dataframe = rules_dataframe.sort_values("lift",ascending = False)
    recommendation_list = []
    for i, j in enumerate(sorted_dataframe["antecedents"]):
        j = int(j)
        if product == j:
            recommendation_list.append(sorted_dataframe.iloc[i,1])
            if len(recommendation_list) == rec_count:
                return recommendation_list
                
                
            
```

```python
arl_recommender(rules, 2_0,4)
```
